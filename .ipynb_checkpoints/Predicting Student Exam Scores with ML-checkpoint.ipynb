### FEATURE ENGINEERING

```python
from sklearn.model_selection import train_test_split, GridSearchCV # GridSearchCV is for parameter tuning
from sklearn.preprocessing import LabelEncoder # LE is for transforming categorical features to numeric
from sklearn.metrics import mean_squared_error, r2_score # These are for model evaluation
from sklearn.linear_model import LinearRegression # For modeling 
from sklearn.ensemble import RandomForestRegressor # For modeling 
from sklearn.tree import DecisionTreeRegressor # For modeling 
```

```python
# Lets select the features that we going to use for our model
df.columns
```

```python
df.head(2)
```

```python
# We not gonna take all the columns, instead we going to select five columns to use on the model 
features = ["study_hours_per_day", "attendance_percentage", "mental_health_rating", "sleep_hours", "part_time_job"]
```

```python
# Now the feature is ready, lets go for the exam score
target = "exam_score"
```

```python
df_model = df[features + [target]].copy() # this is the dataframe for model
```

```python
# In this df, we have part_time_job in categorical, we need to use encoding technique for turn this into a numeric one.
```

```python
le = LabelEncoder()
```

```python
# This label encoder will take NO for 0, YES for 1 --> because in alphabetical order: N comes before Y, hence N is 0 and Y is 1
df_model["part_time_job"] = le.fit_transform(df_model["part_time_job"])
```

```python
# Lets call the dataframe to see the part_time_job column values
df_model
```

```python
# Normally scalling can be a good option but we are going to go with treebased models. So, we are not going to do any kind of scaling operation in here
```

```python
# Now we gonna split the features and label
X = df_model[features]
```

```python
y = df_model[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
len(y_test)
```

```python
len(y_train)
```

```python
# Here we are going to use two treebased models which are decision tree and random forest & then gonna parameter tuning for them
models = {
    "LinearRegression" : {
        "model" : LinearRegression(),
        "params" : {}
    },
    "DescisionTree" : {
        "model" : DecisionTreeRegressor(),
        "params" : {"max_depth" : [3, 5, 10], "min_samples_split" : [2, 5]}
    },
    "RandomForest" : {
         "model" : RandomForestRegressor(),
         "params": {"n_estimators" : [50, 100], "max_depth": [5, 10]}
    }
}
```

```python
# Lets create an empty list with the name of best_models
best_models = []
```

```python
for name, config in models.items():
    print(f"Training {name}")

    grid = GridSearchCV(config["model"], config["params"], cv = 5, scoring = "neg_mean_squared_error")
    grid.fit(X_train, y_train)

    y_pred = grid.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    best_models.append({
        "model" : name,
        "best_params" : grid.best_params_,
        "rmse" : rmse,
        "r2" : r2
    })
```

```python
# So here we are going to train our models & its completed in the above step
best_models # let see the best models list right now
```

```python
# For a better look of best_models
results_df = pd.DataFrame(best_models)
```

```python
# Lets call the results dataframe in here
# sort it by rmse (root mean square error)
results_df.sort_values(by = "rmse")
```

### Interpretation
- We are having LinearRegression at the first place --> which is our winner for this model
- Followed that RandomForest with the max depth of 10 & no. of estimators as 50
- After that, DecisionTree with the max depth of 5 & minimum sample split of 2
